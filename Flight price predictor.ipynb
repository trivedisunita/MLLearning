#Exploratory Data Analysis EDA & Feature Engineering

```python
#step--2
#Exploratory Data Analysis EDA & Feature Engineering

#convert time format in duration column in one single format.
#Duration:
def convert_duration(duration):
    if len(duration.split()) ==2:
        hours = int(duration.split()[0][: -1])
        mins = int(duration.split()[1][: -1])
        return hours*60 + mins
    else:
        return int(duration[: -1])*60
```

```python
data['Duration'] =  data['Duration'].apply(convert_duration)
data.head()
```

```python
##change departure time/arrival time in date format
#ARRIVAL TIME/ DEPARTURE TIME
data['Dep_Time'] = pd.to_datetime(data['Dep_Time'])
data['Arrival_Time']= pd.to_datetime(data['Arrival_Time'])
```

```python
#Extract hours and Mins in Dep. and Arrival time.
data['Dep_Time_Hrs'] = data['Dep_Time'].dt.hour
data['Dep_Time_Mins'] = data['Dep_Time'].dt.minute
data['Arrival_Time_Hrs'] = data['Arrival_Time'].dt.hour
data['Arrival_Time_Mins'] = data['Arrival_Time'].dt.minute
data.head()
```

```python
#As we do not need departure time and arrival time so we should drop these column in dataset.
data.drop(['Dep_Time', 'Arrival_Time'],axis=1, inplace=True)
```

# Apply feature engineering on Date of Journey

```python
data['Date_of_Journey'] = pd.to_datetime(data['Date_of_Journey'])
data.head()
```

```python
# to check  unique value in date of journey column
data['Date_of_Journey'].dt.year.unique()
```
